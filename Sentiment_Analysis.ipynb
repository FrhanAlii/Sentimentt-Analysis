```python
import pandas as pd
df = pd.read_csv('/content/setiment_analysis.csv')
df.head()
```

```python
df = df[['sentence']]
print(df.head())
```

```python
# Check the column names of the DataFrame
print(df.columns)
```

```python
nltk.download('punkt')
df['sentences'] = df['sentence'].apply(lambda text: sent_tokenize(text))
print(df[['sentence', 'sentences']].head())
```

```python
# from nltk.tokenize import word_tokenize

# Tokenize each sentence into words
df['words'] = df['sentences'].apply(lambda sentences: [word_tokenize(sentence) for sentence in sentences])

# Display the words column
print(df[['sentences', 'words']].head())
```

```python
from nltk import pos_tag

# Assign POS tags to each word
df['pos_tags'] = df['words'].apply(lambda words: [pos_tag(word) for word in words])

# Display POS tags for the first row
print(df[['sentences', 'pos_tags']].head())
```

```python
from nltk import ne_chunk

# Apply NER tagging
df['ner_tags'] = df['words'].apply(lambda words: [ne_chunk(pos_tag(word)) for word in words])

# Display NER tags
print(df[['sentences', 'ner_tags']].head())
```

```python
from textblob import TextBlob

# Function to calculate sentiment for each sentence
def get_sentiment(sentences):
    sentiment_scores = []
    for sentence in sentences:
        sentiment_scores.append(TextBlob(sentence).sentiment.polarity)
    return sentiment_scores

# Apply the sentiment analysis function to each row in the 'sentences' column
df['textblob_sentiment'] = df['sentences'].apply(get_sentiment)

# Display the resulting sentiment values
print(df[['sentence', 'sentences', 'textblob_sentiment']].head())
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
df['vader_sentiment'] = df['sentences'].apply(lambda text: analyzer.polarity_scores(text)['compound'])

# Display sentiment values
print(df[['sentences', 'vader_sentiment']].head())
```

```python
print(df.columns)
```

```python

df['textblob_sentiment'] = df['textblob_sentiment'].apply(lambda x: sum(x) / len(x) if isinstance(x, list) else x)
df['vader_sentiment'] = df['vader_sentiment'].apply(lambda x: sum(x) / len(x) if isinstance(x, list) else x)

df['average_sentiment'] = df[['textblob_sentiment', 'vader_sentiment']].mean(axis=1)

print(df[['sentence', 'textblob_sentiment', 'vader_sentiment', 'average_sentiment']].head())
```

```python
from transformers import pipeline

# Load pre-trained sentiment analysis pipeline
classifier = pipeline('sentiment-analysis')

# Apply the transformer model for sentiment analysis
df['transformer_sentiment'] = df['sentences'].apply(lambda text: classifier(text)[0]['label'])

# Print transformer sentiment values
print(df[['sentences', 'transformer_sentiment']].head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score


def categorize_sentiment(polarity):
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'


df['sentiment'] = df['textblob_sentiment'].apply(categorize_sentiment)


X_train, X_test, y_train, y_test = train_test_split(df['sentence'], df['sentiment'], test_size=0.2, random_state=42)


vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train a Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# Predict the sentiment for the test set
y_pred = model.predict(X_test_tfidf)

# Print the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Machine Learning model accuracy(textblob_sentiment): {accuracy}')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Function to categorize sentiment based on polarity
def categorize_sentiment(polarity):
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'


df = df.dropna(subset=['sentence', 'vader_sentiment'])

df['sentiment'] = df['vader_sentiment'].apply(categorize_sentiment)


X_train, X_test, y_train, y_test = train_test_split(df['sentence'], df['sentiment'], test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)


model = RandomForestClassifier(n_estimators=100, random_state=42)  # You can adjust the number of estimators
model.fit(X_train_tfidf, y_train)

# Predict the sentiment for the test set
y_pred = model.predict(X_test_tfidf)

# Print the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Machine Learning model accuracy (Random Forest): {accuracy :.2f}')
```
